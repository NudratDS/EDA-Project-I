```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("movie_profit.csv")
```

```python
data.columns
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
data.dropna(inplace = True)
```

```python
data.duplicated().sum()
```

```python
data.describe()
```

```python
data.sort_values("production_budget" , ascending = False , inplace = True)
```

```python
data["genre"].value_counts()
```

```python
plt.hist(data["worldwide_gross"])
plt.title("Ditribution of World Gross")
plt.xlabel("World Gross")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.bar(data["release_date"].head(),data["domestic_gross"].head())
plt.title("Release Date Domestic Gross")
plt.xlabel("Release Date")
plt.ylabel("Domestic Gross")
plt.show()
```

```python
data.groupby("genre")["production_budget"].sum().plot(kind = "bar")
plt.title("Production Budget by Genre")
plt.xlabel("Genre")
plt.ylabel("Production Budget")
plt.show()
```

```python
data.groupby(data["release_date"].str.slice(6,8))["worldwide_gross"].mean().plot(kind = "line")
plt.title("Mean worldwide Gross Over Year")
plt.xlabel("Release Year")
plt.ylabel("Mean Worldwide Gross")
plt.show()
```

```python
data.groupby("mpaa_rating")["domestic_gross"].sum().plot(kind = "pie")
plt.title("Total Domestic Gross by Rating")
plt.xlabel("Release Date")
plt.ylabel("")
plt.show()

```
